```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('KPMG newCustomerList dataset.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
dd=df['DOB']
dm=dd.mode()
dm.mean()
```

```python
du=df['DOB']
nw=[]
for i in do:
    k=str(i)
    nw.append(k)
sq=[]
nw
for j in nw:
    if j == 'NaT':
        sq.append('1971')
    else:
        sq.append(j[0:4])
dt=[]
for q in sq:
    we=int(q)
    dt.append(we)
```

```python
gn=[]
for t in dt:
    gn.append(2021-t)
```

```python
gh=df['DOB']
fn=df['DOB']
s=0
for i in fn:
    gh = gh.replace([i], [gn[s]])
    s = s+1
s=0
for i in fn:
    df['DOB'] = df['DOB'].replace([i], [gn[s]])
    s = s+1
df=df.rename(columns={"DOB":"AGE"})
df
```

```python
df[['first_name']]=df[['first_name']].fillna(value='Duffill')
f=df
u=f[f['job_industry_category'].isnull()]
u
```

```python
gt=f[df['AGE'] <= 38]
fi=gt['past_3_years_bike_related_purchases'].sum()
```

```python
ran=[]
di=[[0,19],[19,38],[38,44],[44,50],[50,57],[57,64],[64,74],[74,84]]
for d in di:
    gt=df[df['AGE'] <= d[1]]
    gt=gt[gt['AGE'] > d[0]]
    fi=gt['past_3_years_bike_related_purchases'].sum()
    ran.append(fi)
ran
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ra=ran
age=['0-19','20-38','39-44','44-50','51-57','57-64','65-74','75-84']
plt.ylabel('sumes_past_three_years_bike_related_purchas')
plt.xlabel('Age range')
plt.bar(age,ra)
```
